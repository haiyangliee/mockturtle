### Q4 Technical Deep Dive (15 Points):
- Describe the spatial architecture of Eyeriss in your own words.
- How does the Eyeriss architecture optimize energy efficiency in dataflow for CNNs?
- What are the main challenges in designing an energy-efficient architecture for CNNs, and how does Eyeriss tackle these challenges?

Eyeriss features a spatial array consisting of 168 independent Processing Elements arranged in a 14x12 grid. This architecture acts as a localized community where each PE is equipped with its own "Scratchpad" memory for local data storage. To minimize energy-intensive off-chip communication, a specialized Network-on-Chip facilitates multicast data distribution, allowing PEs to pass data directly to neighbors or receive shared data simultaneously.

Eyeriss optimizes energy efficiency by prioritizing Dataflow over raw processing speed through the following mechanisms:     
Row-Stationary Scheduling: This technique keeps a row of filter weights and a row of image data stationary within a PE's local memory.      
Maximizing Local Reuse: It performs the maximum possible calculations with local data before discarding it, reducing expensive "trips" to external memory.      
Energy-Aware Data Movement: By reusing data at the PE level, it avoids the high energy cost of DRAM access, which is 100x–1000x more expensive than MAC operations.      
Global Buffer Management: A 108KB on-chip global buffer caches data to further bridge the gap between PEs and off-chip DRAM

Challenge: The Data Movement Problem Moving data from off-chip DRAM consumes 100x–1000x more energy than actual MACoperations. Solution: Row-Stationary Dataflow Eyeriss keeps a row of the filter and a row of the image data stationary inside the processing element  local memory to minimize expensive "trips" to main memory.  Challenge: High-Dimensional Data Reuse DCNNs involve complex and varied data reuse patterns (convolutional, filter, and feature map reuse) that change across different layers. Solution: Spatial PE Array A grid of 168 independent PEs with local "Scratchpad" storage provides the flexibility and spatial parallelism required to handle diverse layer parameters.  Challenge: Computational Sparsity Processing "zeros" or inactive neurons in AI models wastes significant power and memory bandwidth. Solution: Zero-Skipping Hardware The architecture identifies zeros and skips both the memory fetch and the multiplication, saving up to 45% of power.  Challenge: Data Supply Latency Compute units often stall and lose efficiency while waiting for high-latency DRAM to deliver data. Solution: Buffet & Decoupled Access The Buffet storage mechanism separates data production from consumption, utilizing automated flow control to hide latency and ensure a steady computation stream

### A7:The authors assume that sparsity (zero-valued data) remains a reliable source of energy savings, which is questionable for modern networks using non-ReLU activations (like Swish or GeLU) that produce fewer zeros. They also assume that the static Row-Stationary mapping can efficiently handle all future CNN shapes, but the 14×12 physical PE array leads to hardware underutilization (fragmentation) when layer dimensions do not align with the array's aspect ratio. Additionally, the reliance on a centralized 108KB Global Buffer assumes that on-chip memory bandwidth will not become a bottleneck as compute density scales significantly beyond their 168-PE design.

Future improvements should address dynamic reconfigurability to support non-convolutional layers, such as the attention mechanisms in Transformers, which lack the sliding-window reuse RS dataflow exploits. Another direction is the integration of In-Memory Computing  to eliminate the need for any data movement between the PE and local scratchpads, further reducing the 1x–2x energy cost of local RF/PE access. Furthermore, automating the compiler orchestration for the "Buffet" mechanism would reduce the high manual programming effort required to manage explicit data synchronization.

Eyeriss is significantly more efficient than Weight Stationary or Output Stationary architectures because it minimizes movement for all data types simultaneously rather than just one. While a standard CPU/GPU relies on energy-heavy, implicit cache hierarchies with tag-matching overhead, Eyeriss uses explicitly managed Buffets to achieve predictable, low-power data delivery. Compared to the Google TPU, which uses a large systolic array for rigid matrix multiplication, Eyeriss offers finer-grained flexibility through its independent PE grid and multicast Network-on-Chip, making it better suited for the diverse, irregular workloads of edge devices.
